```python
gender_sales = data.groupby('Gender')['Amount'].sum().reset_index()
ax = sns.barplot(x = 'Gender', y ='Amount', data = gender_sales,hue = 'Gender',palette = 'magma')
plt.title("Sales by Gender")
for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}')
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

From the above chart we can see that most of the customers are females and even the spending power of females are higher than males.

```python
age_group_sales = data.groupby('Age Group')['Amount'].sum().reset_index()
ax = sns.barplot(x = 'Age Group',y ='Amount',data = age_group_sales,hue = 'Age Group',palette = 'magma')
plt.title("Sales by Age Group")
for container in ax.containers:
    ax.bar_label(container,fmt = '{:,.0f}')
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

From the above graphs we can see that most of the customers are of age group between 26-35 years.

```python
state_wise_sales = data.groupby('State')['Amount'].sum().sort_values(ascending = False).head().reset_index()
ax = sns.barplot(x = 'State',y = 'Amount',data = state_wise_sales,hue = 'State',palette = 'magma')
plt.title("Top 5 States wise sales")
plt.xticks(rotation = 90)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

From the above chart we can see the top 5 states with most sales, showing that Uttar Pradesh is the highest followed by Maharastra, Karnataka.

```python
gags = data.groupby(['Age Group','Gender'])['Amount'].sum().reset_index()
sns.barplot(x = 'Age Group', y = 'Amount', hue = 'Gender',data = gags,palette = 'magma')
plt.title("Sales distribution by Age Group and Gender")
plt.show()
```

The above chart tells us that females are still higher in numbers than males and has more purchasing power between the Age Group 26-35 and Gender.

```python
state_gender_sales = data.groupby(['State','Gender'])['Amount'].sum().sort_values(ascending = False).reset_index()
sns.barplot(x = 'State',y = 'Amount',hue='Gender',data = state_gender_sales,palette = 'magma')
plt.title("Sales distribution by State and Gender.")
plt.xticks(rotation = 90)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

From the above chart, again females are the most dominant in contributing to the sales compared to males. Not only in U.P but also in all states females are higher in numbers than males.

```python
occ_sales = data.groupby('Occupation')['Amount'].sum().sort_values(ascending = False).reset_index().head()
sns.barplot(x = 'Occupation',y='Amount',hue = 'Occupation',palette = 'magma',data = occ_sales)
plt.title("Sales by Occupation")
plt.xticks(rotation = 90)
plt.show()
```

From the above chart, we can see the top 5 sales by occupation.Clearly, we can see that people working in IT Sector are having higher spendings followed by Healthcare and Aviation.

```python
product_sales = data.groupby('Product_Category')['Amount'].sum().sort_values(ascending = False).reset_index().head()
sns.barplot(x = 'Product_Category', y = 'Amount',data = product_sales,hue = 'Product_Category',palette = 'magma')
plt.title("Most purchased product categories")
plt.xlabel("Products")
plt.xticks(rotation = 90)
plt.show()
```

From the above chart, we can see that the top 5 most purchased product categories, which shows that Food, Clothing & Apparel , Electronics & Gadgets are the most purchased products.
